```python
import os
import math
import PyGnuplot as gp

def GnuPlot(x, y1, y2, y3, y4, y5, label, min_y, max_y):
    y_tics=""
    for i in range(min_y, max_y+1):
        y_tics+=('''"{}" {},''').format(2**i, i)
    gp.s([x, y1], 'Default')
    gp.s([x, y2], 'Direct')
    gp.s([x, y3], 'Non blocked')
    gp.s([x, y4], 'Ring')
    gp.s([x, y5], 'Ring non blocked')
    
    cmd = '''gnuplot -e "label='{}'" -e "min_y='{}'" -e "max_y='{}'" -e "y_tics='{}'" Plot.p'''.format(label, min_y, max_y, y_tics[-1])
    os.system(cmd)
    

def LogGnuPlot(x, y1, y2, y3, y4, y5, label, min_y, max_y):
    log_x=[math.log2(i) for i in x]
    log_y1=[math.log2(i) for i in y1]
    log_y2=[math.log2(i) for i in y2]
    log_y3=[math.log2(i) for i in y3]
    log_y4=[math.log2(i) for i in y4]
    log_y5=[math.log2(i) for i in y5]
    y_tics=""
    for i in range(min_y, max_y+1):
        y_tics+=('''"{}" {},''').format(2**i, i)
    gp.s([log_x, log_y1], 'Default')
    gp.s([log_x, log_y2], 'Direct')
    gp.s([log_x, log_y3], 'Non blocked')
    gp.s([log_x, log_y4], 'Ring')
    gp.s([log_x, log_y5], 'Ring non blocked')
    cmd = '''gnuplot -e "label='{}'" -e "min_y='{}'" -e "max_y='{}'" -e "y_tics='{}'" LogPlot.p'''.format(label, min_y, max_y, y_tics[-1])
    os.system(cmd)
    
# GnuPlot(MessageSizes, BoringSSL_measured, BoringSSL_predicted, 'e_b_comp', -3, 14)
```

# Hockney Model Errors

```python
def ErrPlot(y1, y2, y3, label):
    p_y1 = [i*100 for i in y1]
    p_y2 = [i*100 for i in y2]
    p_y3 = [i*100 for i in y3]
    x = [i*2 for i in range(10)]
    gp.s([x, p_y1], 'BoringSSL')
    gp.s([x, p_y2], 'Libsodium')
    gp.s([x, p_y3], 'CryptoPP')
    
    cmd = '''gnuplot -e "label='{}'" ErrPlot.p'''.format(label)
    
    
    os.system(cmd)
```

## Errors

```python
def ConcurrentErrPlot(y1, y2, y4, y8, label):
    p_y1 = [i*100 for i in y1]
    p_y2 = [i*100 for i in y2]
    p_y4 = [i*100 for i in y4]
    p_y8 = [i*100 for i in y8]
    x = [i*2 for i in range(10)]
    gp.s([x, p_y1], '1T')
    gp.s([x, p_y2], '2T')
    gp.s([x, p_y4], '4T')
    gp.s([x, p_y8], '8T')
    
    cmd = '''gnuplot -e "label='{}'" ConcurrentErrPlot.p'''.format(label)
    
    
    os.system(cmd)
```

```python
mp_data = pd.read_excel('../PingPongData/PingPongPerformanceModel.xlsx', sheet_name='Concurrent_Recomputed')

E1 = list(mp_data['Unnamed: 11'][2:12])
E2 = list(mp_data['Unnamed: 12'][2:12])
E4 = list(mp_data['Unnamed: 13'][2:12])
E8 = list(mp_data['Unnamed: 14'][2:12])


I1 = list(mp_data['Unnamed: 15'][2:12])
I2 = list(mp_data['Unnamed: 16'][2:12])
I4 = list(mp_data['Unnamed: 17'][2:12])
I8 = list(mp_data['Unnamed: 18'][2:12])


ConcurrentErrPlot(E1, E2, E4, E8, 'Concurrent_Error_Ethernet')
ConcurrentErrPlot(I1, I2, I4, I8, 'Concurrent_Error_IB')

```

```python
import numpy as np
import matplotlib.font_manager as font_manager


Message_sizes = [1, 16, 256, 1024, 4096, 16384, 65536, 262144, 1048576, 2097152]
Comm_pairs = [1, 2, 4, 8]

Eth_data = np.random.rand(10, 4)
IB_data = np.random.rand(10, 4)
for i in range(10):
    Eth_data[i]=[E1[i], E2[i], E4[i], E8[i]]
    IB_data[i]=[I1[i], I2[i], I4[i], I8[i]]
    
font_prop = font_manager.FontProperties(size=20)

fig = plt.figure(figsize=(15, 20))
sp1 = fig.add_subplot(211)
sp2 = fig.add_subplot(212)

h1 = sp1.contourf(Eth_data, cmap='jet', levels=1000)
cbar1 = fig.colorbar(h1, ax=sp1)
cbar1.ax.tick_params(labelsize=18) 
sp1.set_xlabel('number of pairs communicating', fontproperties=font_prop)
sp1.set_ylabel('message size (B)', fontproperties=font_prop)
sp1.set_xticks([x for x in range(4)])
sp1.set_xticklabels(Comm_pairs)
sp1.set_yticks([x for x in range(10)])
sp1.set_yticklabels(Message_sizes)
sp1.tick_params( labelsize=18)


h2 = sp2.contourf(IB_data, cmap='jet', levels=1000)
cbar2 = fig.colorbar(h2, ax=sp2)
cbar2.ax.tick_params(labelsize=18) 
sp2.set_xlabel('number of pairs communicating', fontproperties=font_prop)
sp2.set_ylabel('message size (B)', fontproperties=font_prop)
sp2.set_xticks([x for x in range(4)])
sp2.set_xticklabels(Comm_pairs)
sp2.set_yticks([x for x in range(10)])
sp2.set_yticklabels(Message_sizes)
sp2.tick_params( labelsize=18)

```
