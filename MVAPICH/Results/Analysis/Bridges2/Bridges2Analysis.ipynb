# Bar Plots

```python
def export_data(data, subsets, MessageSizes, caption):
    bars = len(data)
    offsets = [1, 1, 1000, 1000]
    for i in range(len(subsets)):
        with open(caption+'_bar_data_'+str(i)+'.dat', 'w') as output:
            for k in (subsets[i]):
                new_line = str(MessageSizes[k])
                for j in range(bars):
                    new_line +=  " " + str(data[j][k]/offsets[i])
                    if j == bars-1:
                        output.write(new_line)
                        output.write("\n")





def bar_plot(idx, subsets, label, caption):
    types = ['Small', 'Med_S', 'Med_L', 'Large']
    if 'unenc' in caption:
        plotter = 'Bar_Plot_'
    else:
        plotter = 'Enc_Bar_Plot_'
    cmd = '''gnuplot -e "label='{}'" -e "file='{}'" {}{}.p'''.format(label+str(idx), caption+'_bar_data_'+str(idx)+'.dat',plotter, types[idx])
    os.system(cmd)
```

```python
def parse_alignment(file):
    procs = []
    l = [[] for x in range(16)]
    with open(file, 'r') as inp:
        for line in inp:
            if 'Hello' in line:
                toks = line.strip().split()
                proc_ID = int(toks[4][1:4])
                if proc_ID not in procs:
                    procs.append(proc_ID)
                idx = procs.index(proc_ID)
                l[idx].append(int(toks[6]))
                
                
                
    print(procs)
    for ll in l:
        ll= sorted(ll)
        print(ll)
        print()
        print(ll[63]-ll[0])
        print('---')
```
